# Khám phá dữ liệu từ file CSV

Notebook này sẽ thực hiện các bước khám phá dữ liệu (EDA) cho file `data_csv.csv` trong thư mục `data/`. Các bước bao gồm:
- Đọc dữ liệu
- Thống kê tổng quan
- Kiểm tra missing values
- Phân tích từng cột
- Visualize dữ liệu nếu cần thiết

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu từ file CSV
file_path = '../data/data_csv.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# Thông tin tổng quan về dữ liệu
print('Shape:', df.shape)
df.info()
df.describe(include='all')
```

```python
# Kiểm tra missing values
missing = df.isnull().sum()
missing = missing[missing > 0]
if not missing.empty:
    print('Các cột có missing values:')
    print(missing)
else:
    print('Không có missing values!')
```

```python
df.info()
```

```python
df.describe()
```

```python
# Phân tích từng cột: kiểu dữ liệu, số lượng giá trị unique, top value
for col in df.columns:
    print(f'--- {col} ---')
    print('Kiểu dữ liệu:', df[col].dtype)
    print('Số lượng giá trị unique:', df[col].nunique())
    print('Top 5 giá trị phổ biến:')
    print(df[col].value_counts().head())
    print('---\n')
```

```python
# Visualize phân phối các biến số (numeric)
numeric_cols = df.select_dtypes(include=['number']).columns
if len(numeric_cols) > 0:
    df[numeric_cols].hist(figsize=(15, 8), bins=30)
    plt.suptitle('Phân phối các biến số')
    plt.show()
else:
    print('Không có biến số dạng numeric để visualize!')
```

```python
# Visualize mối quan hệ giữa các biến số (nếu có nhiều biến số)
if len(numeric_cols) > 1:
    sns.pairplot(df[numeric_cols])
    plt.suptitle('Pairplot các biến số', y=1.02)
    plt.show()
```

```python
# Visualize các biến phân loại (categorical)
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Phân phối giá trị của {col}')
    plt.xlabel(col)
    plt.ylabel('Số lượng')
    plt.show()
```
